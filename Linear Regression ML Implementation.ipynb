```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(Y_test,Y_pred_lr))
print(mean_squared_error(Y_test,Y_pred_lr))
print(np.sqrt(mean_squared_error(Y_test,Y_pred_lr)))
```

### R square and adjusted R square


Formula

R^2 = 1 - SSR/SST

R^2 = coefficient of determination SSR = sum of squares of residuals SST = total sum of squares

```python
from sklearn.metrics import r2_score

score=r2_score(Y_test,Y_pred_lr)
print(score)
```

Adjusted R2 = 1 – [(1-R2)*(n-1)/(n-k-1)]

where:

R2: The R2 of the model n: The number of observations k: The number of predictor variables

```python

#display adjusted R-squared
1 - (1-score)*(len(Y_test)-1)/(len(Y_test)-X_test.shape[1]-1)
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
rf_model.fit(X_train, Y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
```

```python
# 1. Scatter plot of actual vs predicted
plt.scatter(Y_test, y_pred_rf)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Random Forest - Actual vs Predicted")
plt.show()
```

```python
# 2. Residuals
rf_residuals = Y_test - y_pred_rf
print(rf_residuals.head())
```

```python
# 3. KDE plot of residuals
sns.displot(rf_residuals, kind="kde")
plt.title("Random Forest - Residual Distribution")
plt.show()
```

```python
# 4. Residual vs Prediction
plt.scatter(y_pred_rf, rf_residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Predicted Price")
plt.ylabel("Residuals")
plt.title("Random Forest - Residuals vs Predicted")
plt.show()
```

```python
# 5. Metrics
rf_mae  = mean_absolute_error(Y_test, y_pred_rf)
rf_mse  = mean_squared_error(Y_test, y_pred_rf)
rf_rmse = np.sqrt(rf_mse)
```

```python
print("Random Forest Metrics:")
print(f"MAE  : {rf_mae:.3f}")
print(f"MSE  : {rf_mse:.3f}")
print(f"RMSE : {rf_rmse:.3f}")
```

```python
# 6. R² and Adjusted R²
rf_r2 = r2_score(Y_test, y_pred_rf)
n, k = X_test.shape
rf_adj_r2 = 1 - (1 - rf_r2) * (n - 1) / (n - k - 1)

print(f"R²    : {rf_r2:.3f}")
print(f"Adj R²: {rf_adj_r2:.3f}")
```

```python
# 5. Metrics
xgb_mae  = mean_absolute_error(Y_test, y_pred_xgb)
xgb_mse  = mean_squared_error(Y_test, y_pred_xgb)
xgb_rmse = np.sqrt(xgb_mse)

print("XGBoost Metrics:")
print(f"MAE  : {xgb_mae:.3f}")
print(f"MSE  : {xgb_mse:.3f}")
print(f"RMSE : {xgb_rmse:.3f}")
```

```python
# 6. R² and Adjusted R²
xgb_r2 = r2_score(Y_test, y_pred_xgb)
xgb_adj_r2 = 1 - (1 - xgb_r2) * (n - 1) / (n - k - 1)

print(f"R²    : {xgb_r2:.3f}")
print(f"Adj R²: {xgb_adj_r2:.3f}")
```

### Overall Modal Evaluation

```python
def evaluate_model(name, y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    print(f"{name} - RMSE: {rmse:.4f}, R²: {r2:.4f}")

evaluate_model("Linear Regression", Y_test, Y_pred_lr)
evaluate_model("Random Forest", Y_test, y_pred_rf)
evaluate_model("XGBoost", Y_test, y_pred_xgb)
```
